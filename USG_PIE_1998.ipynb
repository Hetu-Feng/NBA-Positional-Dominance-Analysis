```python
from selenium import webdriver
from pandas import *
import pandas
import numpy as np

#open the stat.nba.com webpage with selenium driver
url = 'https://stats.nba.com/players/advanced/?sort=GP&dir=-1&Season=1998-99&SeasonType=Regular%20Season'

EXE_PATH = '/Users/hetufeng/Downloads/chromedriver'
driver = webdriver.Chrome(executable_path=EXE_PATH)
driver.get(url)
driver.implicitly_wait(100)
```

```python
driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/div[1]/div[1]/div/div/label/select/option[22]').click()
driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/div[1]/div[2]/div/div/label/select/option[2]').click()
driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[3]/div/div/select/option[1]').click()
```

```python
table = driver.find_element_by_class_name('nba-stat-table__overflow')
```

```python

#find player name,and fga. added to list
player_names = []
player_stats = []
stats = []
names = []
usg = []
pie = []
gp = []
minutes = []
usg_98 = []
pie_98=[]

for line_id, lines in enumerate(table.text.split('\n')):
    if line_id == 0:
        column_names = lines.split(' ')[1:]
    else:
        if line_id % 3 == 2:
            player_names.append(lines)
        if line_id % 3 == 0:
            player_stats.append( [i for i in lines.split(' ')] )
        
names = player_names[:400]

stats = player_stats[:400]

for stat in stats:
    usg.append(float(stat[18]))
    pie.append(float(stat[20]))
    gp.append(float(stat[2]))
    minutes.append(float(stat[5]))
    
    
for i in range(len(usg)):
    usg_98.append([names[i],usg[i], gp[i], minutes[i]])
    pie_98.append([names[i],pie[i], gp[i], minutes[i]])
```
